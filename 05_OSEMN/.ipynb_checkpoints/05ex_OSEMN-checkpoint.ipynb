1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
#first version
import random
import datetime

nums=100
filename="simple_data.txt"

with open(filename, "w") as w:
    for n in range(0, nums):
        w.write(str(random.randint(0,10000)))
        w.write(",")
    w.write(str(random.randint(0,10000)))

with open(filename, "r") as f:
    rand_nums = list(map(int, f.read().split(',')))

print(rand_nums[0:int(nums/4)])
```

```python
#second version
import random
import datetime

filename1="simple_data.txt"
rand_matrix=np.random.rand(10)
with open(filename1,"w") as f:
        np.savetxt(f, rand_matrix, fmt='%.3f')
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
rand_matrix=np.random.rand(5,5)
mat=np.matrix(rand_matrix)
with open('data.txt',"w") as f:
    for line in mat:
        np.savetxt(f, line, fmt='%.2f')
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
#input file
fin = open("data.txt", "rt")
#output file to write the result to
fout = open("data.csv", "wt")
#for each line in the input file
for line in fin:
    #read replace the string and write to output file
    fout.write(line.replace(' ', ','))
#close input and output files
fin.close()
fout.close()
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
with open("credit_card.dat", "r") as f:
    for line in f:
        if len(line)<6:
            print("\nDone!")
        else:
            nums=[]
            for i in range(19):
                nums.append(chr(int(line[i*6:(i+1)*6],2)))
        
            card = "".join(nums)        
            print(card)
```
